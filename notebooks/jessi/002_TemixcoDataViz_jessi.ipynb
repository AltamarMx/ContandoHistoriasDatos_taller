```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = "../../data/temixco.csv"

tmx = pd.read_csv(f, index_col=0, parse_dates=True)
```

```python
dia = tmx[["Ig", "To"]].groupby(by=[tmx.index.strftime("%H:%M")]).mean()
```

```python
dia.info()
```

```python
with plt.xkcd():
    dia.plot()
```

```python
def formato_ax(axs):
    for ax in axs:
        ax.legend()
        ax.spines[["right", "top"]].set_visible(False)
```

```python
fig, ax = plt.subplots(2, figsize =(10,4))

ax[0].plot(dia.To, label="To")
ax[1].plot(dia.Ig, label="Ig")

ax[0].set_ylabel("Temperatura[°C]")
ax[1].set_ylabel("Irradiancia[$W/m^2$]")
ax[1].set_xlabel("Tiempo [$:HH:MM$]")

formato_ax(ax)
```
