```python
%matplotlib inline
import matplotlib.pyplot as mp
import pandas as pd
import numpy as np
```

```python
f  = "../data/temixco temperaturas.txt"
nombres = ["tiempo","Tmax","Tmin"]

df = pd.read_csv(f,encoding ="iso-8859-1", skiprows = 18,
                 delim_whitespace = True , usecols =[0,3,4],
                 names = nombres,na_values ="Nulo")
df.tiempo = pd.to_datetime(df.tiempo,format = "%d/%m/%Y")
df.set_index("tiempo", inplace =True)
df
```
