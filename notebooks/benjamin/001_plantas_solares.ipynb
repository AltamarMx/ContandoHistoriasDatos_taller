```python
fig, ax = plt.subplots(figsize=(15,6))

#ps=df['Area instalada (gross), m2'].groupby(by=[df.index]).sum()
ps=df['Installed thermal power , kWth'].groupby(by=[df.index]).sum().cumsum()
ax.set_ylabel('kWth')
ax.plot(ps)
ax.set_title('Accumulate installed thermal power')
plt.show()
```
