```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = "../../data/Plantas termosolares México.xlsx"   #df significa Data Frame
df = pd.read_excel(f)
df

#DataFrame 1
#tengo los datos tal cual están en el archivo de excel
```

```python
años = pd.read_excel(f,index_col=5)
años.sort_index(inplace=True)
años

#DataFrame 2  (bueno, no es que sean datos diferentes, solo ordenados de diferente manera)
#Hago que se orden por el año de inicio de operación
```

```python
df['Installed thermal power , kWth'].sum()

#Hace la suma de toda la potencia térmica instalada con el Data Frame 1
```

```python
años['Installed thermal power , kWth'].groupby(by=[años.index]).sum()

#usando el data frame 2, donde los datos están 
#agrupados por años, hace la suma en cada año
```
