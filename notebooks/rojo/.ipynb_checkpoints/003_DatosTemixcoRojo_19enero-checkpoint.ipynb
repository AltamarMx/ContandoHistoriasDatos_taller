```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
azulier      ="#1A3D6F"
doradoier   ="#C65C25"
```

```python
f = "../../data/temixco.csv"   #df significa Data Frame
tmx = pd.read_csv(f,index_col=0,parse_dates=True)
tmx
```

```python
tmx.groupby(by=[tmx.index])
```

```python
tmx.index.strftime("%H:%M")
```

```python
dia = tmx[["Ig","To"]].groupby(by=[tmx.index.strftime("%H:%M")]).mean()
```

```python
dia.plot()
```

### Definiendo el formato "general" de mis figuras

```python


def formato_ax(axs):
    for ax in axs:
#        ax.legend() #activa que las graficas tenga leyendas
        ax.spines[["right","top"]].set_visible(False) #elimina el margen de arriba, y de la derecha
#        ax.grid()
        ax.spines["bottom"].set_color("grey")
        ax.spines["left"].set_color("grey")    
    
    
```

### Figura 1

```python
#estÃ¡ recibiendo dos objetos (que se estÃ¡n iterando) por eso estÃ¡n los []
fig, ax = plt.subplots(2,figsize=(10,4))

ax[0].plot(dia.To, label="To")
ax[1].plot(dia.Ig, label="Ig")


formato_ax(ax)
#formato__ax(ax[1])
```

### Figura 2

```python
#estÃ¡ recibiendo dos objetos tambiÃ©n, pero yo los "defino" y no se iteran
fig, (ax1,ax2) = plt.subplots(2,figsize=(10,4))

ax1.plot(dia.To, label="To")
ax2.plot(dia.Ig, label="Ig")


formato_ax(ax)
#formato__ax(ax[1])
```

### Figura 3

```python
#MISMA QUE LA FIGURA 1, PERO  CON MÃS FORMATO
#estÃ¡ recibiendo dos objetos (que se estÃ¡n iterando) por eso estÃ¡n los []

fig, ax = plt.subplots(2,figsize=(10,4),sharex=True)  #hace que compartan el mismo eje

ax[0].plot(dia.To, label="To",c=azulier, lw=4)  #lw es para el grosor de las lineas
ax[1].plot(dia.Ig, label="Ig",c=doradoier, lw=4)

ax[0].set_ylim(15,30)   #limite del eje Y de la grafica 1

#ax.set_ylabel("Y", rotation=0, labelpad=10)

ax[0].set_ylabel("Temperatura [$^oC$]", labelpad=10)  #etiqueta de eje Y, grafica 1
ax[1].set_ylabel("Irradiancia [$W/m^2$]")  #etiqueta de eje Y, grafica 2
ax[1].set_xlabel("Tiempo [$HH:MM$]")   #etiqueta de eje X para ambas graficas

formato_ax(ax)
#formato__ax(ax[1])
```
